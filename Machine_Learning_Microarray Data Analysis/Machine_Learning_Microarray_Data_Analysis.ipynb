```r

#
# =====================================================================
#                      Microarray Data Analysis_Genomics
# =====================================================================


# --------------------------
# Load Required Packages
# --------------------------
# Each library below supports a specific step in our ML workflow
install.packages("downloader")
library(downloader)
install.packages("caret")
library(caret)         # End-to-end ML workflow (preprocessing, training, validation)
install.packages("RANN")
library(RANN)          # KNN-based imputation of missing values
install.packages("randomForest")
library(randomForest)  # Random Forest algorithm
install.packages("Boruta")
library(Boruta)        # Wrapper for feature selection
install.packages("kernlab")
library(kernlab)       # SVM algorithms
library(pROC)          # ROC curve and AUC calculation
library(ggplot2)       # Visualization (for plots, bar charts, ROC etc.)

################################################################################
raw_url_Colorectal<-"https://raw.githubusercontent.com/AI-Biotechnology-Bioinformatics/AI_and_Omics_Research_Internship_2025/refs/heads/main/Datasets/Colorectal_Cancer.csv"
file_name<-"Colorectal_Cancer.csv"
download(url=raw_url_Colorectal,destfile=file_name)
data_Colorectal<-read.csv("Colorectal_Cancer.csv",row.names = 1)

raw_url_brief_Colorectal<-"https://raw.githubusercontent.com/AI-Biotechnology-Bioinformatics/AI_and_Omics_Research_Internship_2025/refs/heads/main/Datasets/Brief_Colorectal_Cancer.csv"
file_name<-"Brief_Colorectal_Cancer.csv"
download(url=raw_url_brief_Colorectal,destfile=file_name)
data_Brief_Colorectal<-read.csv("Brief_Colorectal_Cancer.csv",row.names = 1)

#row.names = 1 means:

#“Use the first column of the CSV file as the row names of the data frame,
# not as a normal variable column.”

################################################################################
dim(data_Colorectal)
str(data_Colorectal)
dim(data_Brief_Colorectal)
str(data_Brief_Colorectal)

# Many expression datasets are skewed (some genes have very large values).
# Log transformation helps reduce skewness and stabilize variance.
boxplot(data_Colorectal[1:30],
        outline = FALSE,
        col = "skyblue",
        main = "Before Log Transformation")

par(mfrow = c(1, 2))
#par() changes plotting parameters.
#mfrow means “multi-figure layout by rows”.
#c(1, 2) means 1 row, 2 columns.

hist(data_Colorectal[, 18], main = "Before Log Transform", col = "skyblue")
hist(log10(data_Colorectal + 1)[, 18], main = "After Log Transform",
     col = "coral")

data_Colorectal_log <- log10(data_Colorectal + 1)  # +1 avoids log(0) which is undefined

#  ---- Step 2. Transpose Data (Samples = Rows, Genes = Columns)  ----
# ---------------------------------------------------------------------
# ML algorithms expect data in the format: rows = samples, columns = features.
data_Colorectal_t <- as.data.frame(t(data_Colorectal_log))

#  ---- Step 3. Feature Filtering  ----
# -------------------------------------
# Some genes show almost no variation across samples - these carry little information.
# Removing them reduces computational load and improves learning.
# nzv = near zero variance



nzv <- preProcess(data_Colorectal_t, method = "nzv", uniqueCut = 15)
data_Colorectal_t <- predict(nzv, data_Colorectal_t)



# Optionally, you can filter by variance or correlation.
# to keep only the top variable genes (e.g., top 500 or 1000) that avoid redundancy.

# ------------------------------------------------------------------------------
#  ---- Step 4. Center and Scale Features
# ----------------------------------------
# For more numeric stability, we van center and scale our data
# Centering subtracts the mean; scaling divides by standard deviation.
# Ensures all genes contribute equally to the model (important for SVM/ANN).
#What actually changes inside data_Colorectal_t?

#For each column separately (Gene1, Gene2, …):
#R computes the mean and standard deviation of that column.
#Then it replaces every value by:

#new value= old value−column mean/column sd

#So for every gene/feature column:

#1.Before: raw intensities (maybe 0.3, 12, 500, …)

#2.After: transformed values with :average ≈ 0,spread (sd) ≈ 1

#Tiny numeric example
#Imagine data_Colorectal_t has one column:
#[2, 4, 6]
#mean = 4
#sd ≈ 2

#After centering & scaling:
#(2 − 4) / 2 = −1
#(4 − 4) / 2 = 0
#(6 − 4) / 2 = 1

#So the column becomes [-1, 0, 1].
#Your real dataset does this for every gene column.

#Why store it in process?
#process just remembers the means and sds for each column.
#Then predict(process, data_Colorectal_t) uses those to transform the data.
#Later you can use the same process object to transform a test set in exactly the same way.

#So in short:

#That code takes your dataset data_Colorectal_t and turns every feature column
#into “z-scores” (centered at 0, scaled to sd 1), and overwrites data_Colorectal_t with these standardized values.

#1)Make all features comparable:
#In omics, one gene can have values like 0.01–1, another 1000–100000.If you don’t scale:
#Algorithms that use distances or dot products (SVM, k-means, PCA, neural nets, etc.) are dominated by the large-scale genes.
#Small‐scale genes are almost “invisible” to the model.
#After centering & scaling:
#Every gene has mean ≈ 0 and sd ≈ 1
#So each gene contributes roughly equally.

#2)Help optimization / training
#Many ML algorithms use gradient descent or similar optimization.
#If features are on very different scales:
#Training can be slow
#Convergence can be unstable or poor
#Standardized data makes the optimization landscape more “nice” → models train faster and more reliably.

#3)Some methods mathematically assume it
#Methods like:PCA,k-means,Regularized regression (lasso, ridge),Many distance-based methods
#really expect data to be standardized. Without it, the interpretation of components/clusters/coefficients can be misleading.

process <- preProcess(data_Colorectal_t, method = c("center", "scale"))
data_Colorectal_t <- predict(process, data_Colorectal_t)

# -------------------------------------------
#  ---- Step 5. Handle Missing Values  ----
# -------------------------------------------
# Missing values are common in omics data.
# We either remove or impute them.

anyNA(data_Colorectal_t)
sum(is.na(data_Colorectal_t))

# Option 1: Remove missing values either (samples or features)- simple but can lose data.
# Option 2: Impute missing values (recommended).OK
# Simply impute missing values by mean or median values
# KNN imputation is another method that is more precise than mean and median.
# KNN imputation replaces missing values using the mean of the k nearest samples.
# KNN find samples that are momst similar to sample vectors wit NA values.
# Then the algorithm(KNN) averaghes the non-NA values & replace them in NA values.
knnImpute <- preProcess(data_Colorectal_t, method="knnImpute")
data_Colorectal_t <- predict(knnImpute, data_Colorectal_t)
sum(is.na(data_Colorectal_t))  # Check again to confirm all missing values are handled.


# -------------------------------------
#  ---- Step 6. Add Class Labels  ----
# -------------------------------------
# Machine Learning requires both input (X) and output (Y).
# Here, Y = class labels (e.g., cancer vs normal).
groups <- factor(data_Brief_Colorectal$Characteristics.sampling.site.,
                 levels = c("cancer", "normal"),
                 labels = c(1, 0))

class(groups)
levels(groups)
table(groups)

data_Colorectal_t <- cbind(groups, data_Colorectal_t)
View(data_Colorectal_t)


#--------------------------------------
# ---- Step 7. DATA PARTITIONING  ----
#--------------------------------------

# Split data into Training (70%) and Testing (30%) for unbiased evaluation.
set.seed(123)
trainIndex <- createDataPartition(data_Colorectal_t$groups, p = 0.7, list = FALSE)
train_data <- data_Colorectal_t[trainIndex, ]
dim(train_data)

test_data  <- data_Colorectal_t[-trainIndex, ]
dim(test_data)

x_train <- train_data[, -1]  # Gene expression matrix
y_train <- train_data$groups # Class labels

x_test  <- test_data[, -1]
y_test  <- test_data$groups

#-------------------------------------------------------------------------------
#### FEATURE SELECTION ####
#----------------------------#

# The goal is to reduce thousands of genes to a small informative subset
# that carries the most predictive signal for the classification task.

#--------------------------------------------------
#### Method 1: Feature Selection with Boruta ####
#--------------------------------------------------

# Boruta is a wrapper around Random Forest that tests each feature’s importance.
# It keeps only those that significantly contribute to the prediction.

library(Boruta)
library(randomForest)

boruta_result<-Boruta(x=x_train,y=y_train,
                      doTrace = 1)
boruta_result

#fix tentative features
final_boruta<-TentativeRoughFix(boruta_result)
final_boruta

boruta_features<-getSelectedAttributes(final_boruta)
boruta_features
write.csv(boruta_features,"Results_Boruta_Selected_Genes.csv",row.names = FALSE)

#-------------------------------------------------------------------------------
#### Method 2: Feature Selection with RFE (Recursive Feature Elimination) ####
#-------------------------------------------------------------------------------
# RFE iteratively removes least important features and retrains the model.

library(randomForest)
library(caret) #provide RFE fraemwork

rfe_control<-rfeControl(functions = rfFuncs,#random forest
                        method = "cv", # cross validation method
                        number = 3, # 3 fold cross_validation
                        verbose = TRUE)

rfe_result <- rfe(x = x_train,
                  y = y_train,
                  sizes = c(1:5, 10, 50), # we can skipp this step , so RFE try on all features
                  rfeControl = rfe_control)
rfe_result
plot(rfe_result, type = c("g", "o"))
rfe_features <- predictors(rfe_result)
write.csv(rfe_features, "Results_RFE_Selected_Genes.csv", row.names = FALSE)
rfe_features

common_features<-intersect(boruta_features,rfe_features)
common_features
length(common_features)
write.csv(common_features,"Results_Common_Selected_Features(genes).csv")

#-------------------------------------------------------------------------------
#### MODEL TRAINING & EVALUATION ####
#-------------------------------------------------------------------------------

# We’ll compare three classifiers: Random Forest, SVM, and ANN.
# Each learns patterns differently.

set.seed(123)

library(caret)
library(randomForest)
library(pROC)
library(ggplot2)

ctrl <- trainControl(method = "cv", number = 10, verboseIter = TRUE)

#The trainControl function is from the caret package and its job is to specify:
#How the model should perform validation while training.
#verboseIter = TRUE==>"Print a progress message on the screen at each cross-validation iteration."

## ---- BORUTA Feature Set -----------------------------------------------------

#Keep all rows and take just boruta_features as columns

train_boruta<-x_train[,boruta_features]
test_boruta<-x_test[,boruta_features]

# ----------------------------------
# ---- Model 1: Random Forest  ----
# ----------------------------------

RF_boruta <- train(x = train_boruta, y = y_train, method = "rf",
                   importance = TRUE, # after training, let's calculate the importance of each gene as a feature
                   trControl = ctrl)

plot(varImp(RF_boruta), top = 20)  # Visualize top contributing genes

# -----------------------------------------
#  ---- Model 2: SVM (Radial Kernel)  ----
# -----------------------------------------
SVM_boruta<-train(x=train_boruta,
                  y=y_train,
                  method="svmRadial",
                  trControl = ctrl,
                  tuneGrid= data.frame(C=c(0.25,0.5,1)
                                       , sigma= 0.5),
                  prob.model = TRUE)

#C is a control on "how sensitive do you want to be to train errors?"
#big_c ==> wants almost no errors in the train data
#big_c==> over fitting risk
# Then we see which "c" has better function in "cross validation

# what is Sigma?

#SVM with RBF kernel seems to draw soft circles around the points.
#sigma is basically saying:
#"How wide/narrow should these circles be?"
#small sigma  circles too narrow  boundary can get very complex
#large sigma  circles big and soft  smoother boundary


#prob.model = TRUE

#mean:
# “Don’t just say class 0 or 1;
#but give me the probability of each class also

SVM_boruta$bestTune

# --------------------------
#  ---- Model 3: ANN  ----
# --------------------------
ann_boruta <- train(x = train_boruta, y = y_train, method = "nnet",
                    trControl = ctrl,
                    tuneGrid = data.frame(size = 1:2, decay = 0),
                    MaxNWts = 2000)
ann_boruta$bestTune

ann_boruta <- train(
  x = train_boruta,
  y = y_train,
  method = "nnet",
  trControl = ctrl,
  tuneGrid = expand.grid(
    size  = 1:5,            # Test 1,2,3,4,5 hidden neurons
    decay = c(0, 0.001)     # Two values  for decay
  ),
  MaxNWts = 5000
)
ann_boruta$bestTune

# -----------------------------
#  ---- Model Evaluation  ----
# -----------------------------

# Prediction
rf_pred_boruta  <- predict(RF_boruta, newdata = test_boruta)
svm_pred_boruta <- predict(SVM_boruta, newdata = test_boruta)
ann_pred_boruta <- predict(ann_boruta, newdata = test_boruta)

# Confusion Matrix shows model’s classification performance
rf_conf_boruta  <- confusionMatrix(rf_pred_boruta, y_test)
svm_conf_boruta <- confusionMatrix(svm_pred_boruta, y_test)
ann_conf_boruta <- confusionMatrix(ann_pred_boruta, y_test)

# confusion_metrics
rf_conf_boruta<-confusionMatrix(rf_pred_boruta,y_test)
svm_conf_boruta<-confusionMatrix(svm_pred_boruta,y_test)
ann_conf_boruta<-confusionMatrix(ann_pred_boruta,y_test)

rf_conf_boruta
svm_conf_boruta
ann_conf_boruta

#accuracy: how many samples we correctly classified
#sensitivity:how well the model identifies the True Positive class(Cancer)
#specificity : measures how well the model identifies True Negative(Control) samples
#Balanced Accuracy: Helps us when classes are imbalanced(if we have more cancer>ctrl, ctrl samples may get miss classified
# so Balanced Accuracy averages the sensitivity and  specificity so then we have same raw and balance accuracy because we have balance dataset  )


# Compare accuracy
results_boruta <- data.frame(
  Model = c("Random Forest", "SVM", "ANN"),
  Accuracy = c(rf_conf_boruta$overall["Accuracy"],
               svm_conf_boruta$overall["Accuracy"],
               ann_conf_boruta$overall["Accuracy"])
)
print(results_boruta)
table(y_test)

# training error
# 1 - accuracy(model's accuracy)

# ROC & AUC Curves
# ROC curve = True Positive Rate vs False Positive Rate
# AUC = area under ROC; measures model discrimination

#------------------------- -----ROC & AUC --------------------------------------------
rf_prob_boruta  <- predict(RF_boruta, newdata = test_boruta,type = "prob")
svm_prob_boruta <- predict(SVM_boruta, newdata = test_boruta,type = "prob")
ann_prob_boruta <- predict(ann_boruta, newdata = test_boruta,type = "prob")

#If you use type = "raw" (or no type at all)  it gives class labels (e.g. “Cancer”, “Normal”).
#With type = "prob"  it gives probabilities for each class.

head(rf_prob_boruta)

rf_roc_boruta  <- roc(y_test, as.numeric(rf_prob_boruta[, 1]))
svm_roc_boruta <- roc(y_test, as.numeric(svm_prob_boruta[, 1]))
ann_roc_boruta <- roc(y_test, as.numeric(ann_prob_boruta[, 1]))


#### OR ####

rf_roc_boruta <- roc(
  response  = y_test,
  predictor = as.numeric(rf_prob_boruta[, 1]),
  levels    = c(0, 1),   # 0 = control, 1 = case (positive)
  direction = "<"        # The higher the probability,  more towards class 1
)

svm_roc_boruta <- roc(
  response  = y_test,
  predictor = as.numeric(svm_prob_boruta[, 1]),
  levels    = c(0, 1),
  direction = "<"
)

ann_roc_boruta <- roc(
  response  = y_test,
  predictor = as.numeric(ann_prob_boruta[, 1]),
  levels    = c(0, 1),
  direction = "<"
)

rf_auc_boruta <- auc(rf_roc_boruta)
svm_auc_boruta <- auc(svm_roc_boruta)
ann_auc_boruta <- auc(ann_roc_boruta)


rf_auc_boruta; svm_auc_boruta; ann_auc_boruta

#AUC_RF:0.9896
#AUC_SVM:0.965
#AUC_ANN:0.9872

# Plot ROC curves for visual comparison
par(mfrow = c(1, 3))
plot(rf_roc_boruta, col = "black", lwd = 2, main = paste("RF (AUC =", round(rf_auc_boruta, 3), ")"))
plot(svm_roc_boruta, col = "green", lwd = 2, main = paste("SVM (AUC =", round(svm_auc_boruta, 3), ")"))
plot(ann_roc_boruta, col = "blue", lwd = 2, main = paste("ANN (AUC =", round(ann_auc_boruta, 3), ")"))

#=============================================================================================================================

## ---- RFE Feature Set ----


train_rfe <- x_train[, rfe_features]
test_rfe  <- x_test[, rfe_features]

# Train again using RFE features to compare results

# Random Forest
RF_rfe <- train(x = train_rfe, y = y_train, method = "rf",
                importance = TRUE,
                trControl = ctrl)

plot(varImp(RF_boruta), top = 10)  # Visualize top contributing genes

# SVM
svm_rfe <- train(x = train_rfe, y = y_train, method = "svmRadial",
                 trControl = ctrl,
                 tuneGrid= data.frame(C=c(0.25,0.5,1), sigma= 0.5),
                 prob.model = TRUE)

#ANN
ann_rfe <- train(x = train_rfe, y = y_train, method = "nnet",
                 trControl = ctrl,
                 tuneGrid = data.frame(size = 1:2, decay = 0),
                 MaxNWts = 2000)

# Evaluate and compare
rf_pred_rfe  <- predict(RF_rfe, newdata = test_rfe)
svm_pred_rfe <- predict(svm_rfe, newdata = test_rfe)
ann_pred_rfe <- predict(ann_rfe, newdata = test_rfe)

rf_conf_rfe  <- confusionMatrix(rf_pred_rfe, y_test)
svm_conf_rfe <- confusionMatrix(svm_pred_rfe, y_test)
ann_conf_rfe <- confusionMatrix(ann_pred_rfe, y_test)

results_rfe <- data.frame(
  Model = c("Random Forest", "SVM", "ANN"),
  Accuracy = c(rf_conf_rfe$overall["Accuracy"],
               svm_conf_rfe$overall["Accuracy"],
               ann_conf_rfe$overall["Accuracy"])
)
print(results_rfe)

# AUC comparison plot
rf_prob_rfe  <- predict(RF_rfe, newdata = test_rfe, type = "prob")
svm_prob_rfe <- predict(svm_rfe, newdata = test_rfe, type = "prob")
ann_prob_rfe <- predict(ann_rfe, newdata = test_rfe, type = "prob")

rf_roc_rfe  <- roc(y_test, as.numeric(rf_prob_rfe[, 1]))
svm_roc_rfe <- roc(y_test, as.numeric(svm_prob_rfe[, 1]))
ann_roc_rfe <- roc(y_test, as.numeric(ann_prob_rfe[, 1]))

rf_auc_rfe <- auc(rf_roc_rfe)
svm_auc_rfe <- auc(svm_roc_rfe)
ann_auc_rfe <- auc(ann_roc_rfe)


rf_auc_rfe; svm_auc_rfe; ann_auc_rfe

dev.off()
plot(rf_roc_rfe, col = "blue", lwd = 2, main = "ROC Curve - RFE Features")
plot(svm_roc_rfe, col = "red", lwd = 2, add = TRUE)
plot(ann_roc_rfe, col = "green", lwd = 2, add = TRUE)

legend("bottomright",
       legend = c(
         paste("RF (AUC =", round(auc(rf_roc_rfe), 3), ")"),
         paste("SVM (AUC =", round(auc(svm_roc_rfe), 3), ")"),
         paste("ANN (AUC =", round(auc(ann_roc_rfe), 3), ")")
       ),
       col = c("blue", "red", "green"), lwd = 2, bty = "n")
```
